```python
import LAB_FEEC_UNICAMP as LAB
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import glob
import pickle as pk
%matplotlib notebook
```

```python
#Demo Audio Files:
#Audios/pulse.wav
#Audios/speech.wav

#Configurate Initial Parameters
c=343.3

#Initialize Array
arr = LAB.Array(c=c)
#Simulates the audio in four microphones and its reflection on the sample
pos_mics = [0.052,0.51,1.03,1.53] #Position of each microphone
pos_mic_amos = 2 #Distance between microphone and the sample
arr.simulatefromWav('Audios/speech.wav',pos_mics,pos_mic_amos,coefInc=1,coefRef=0.2)

#Plot Mic signals in time
plt.figure('Microphones',figsize=(9,5))
LAB.Plot_Mics_Time(arr)
```

```python
DAS_Spectrum = LAB.applyDAS(arr, src_to_smpl = True)
plt.figure('DAS Inc(Spectrum)',figsize=(9,5))
LAB.plotFreq(arr.fs,DAS_Spectrum)
plt.figure('DAS Inc(Time)',figsize=(9,5))
LAB.plotTime(arr.fs,DAS_Spectrum)
```

```python
Cov_Parts = 3
diagonalLoading = 0 #Adaptative
xmin = 250
xmax = 8000

#Estimates noise spectrum
N = LAB.applyDAS(arr, src_to_smpl = False)
Dls = LAB.DeslocSig(N,LAB.calcManifoldVector(arr, False))
N = np.matrix(Dls)
Rxx = LAB.Beamforming.IterativeCovariance(arr, N, Cov_Parts)
Rxx_inv = LAB.invertRxx(Rxx, diagonalLoading)

MVDR_Inc = LAB.applyMVDR(arr, src_to_smpl = True, Rxx_inv = Rxx_inv)
plt.figure('MVDR Inc(Spectrum)',figsize=(9,5))
LAB.plotFreq(arr.fs,MVDR_Inc)
plt.figure('MVDR Inc(Time)',figsize=(9,5))
LAB.plotTime(arr.fs,MVDR_Inc)
```

```python
Dls = LAB.DeslocSig(MVDR_Inc,LAB.calcManifoldVector(arr, True))
N = np.matrix(Dls)

Rxx_Refl = LAB.IterativeCovariance(arr, N, Cov_Parts)
Rxx_Inv_Refl = LAB.invertRxx(Rxx_Sn_Refl,diagonalLoading)

MVDR_Refl = LAB.applyMVDR(arr, False, Rxx_Sn_Inv_Refl)

#Refl Coef
plt.figure(figsize=(9,5))
plt.title('Coeficiente de ReflexÃ£o')
rate = (np.absolute(MVDR_Refl)/np.absolute(MVDR_Inc))
rate = pd.DataFrame(rate).rolling(300,min_periods=1,center=True).median()._values
plt.xlabel('FrequÃªncia(Hz)')
plt.ylabel('Amplitude')
plt.xlim(xmin,xmax)
plt.semilogx(np.linspace(0,arr.fs/2-1,len(MVDR_Refl)),rate,'b')
```

```python
pasta = 'Simulados'

arquivos = glob.glob("data/"+pasta+"/*.pickle")

MVDR_Incs = []
MVDR_Refls = []
Conds_Rxx_Global = []
Rates_Median = []
fs=0
for a in arquivos:
    with open(a, 'rb') as f:
        T = pk.load(f)
        mI = T['MVDR_Inc']
        mR = T['MVDR_Refl']
        MVDR_Incs.append(mI)
        MVDR_Refls.append(mR)

        Rates_Median.append([1 - (np.absolute(mR[i])/np.absolute(mI[i]))**2 for i in range(0,len(mI))])

        #conds = T['Conds_Rxx_Inc']
        #conds = conds/np.mean(conds)
        #Conds_Rxx_Global.append(conds)

        fs = T['fs']
        f.close()

Rates_Median = np.median(Rates_Median,axis=0)

Rates = pd.DataFrame(Rates_Median).rolling(300,min_periods=1,center=True).median()._values

plt.figure()
plt.title('Coef. AbsorÃ§Ã£o')
plt.xlim(250,8000)
plt.ylim(0,1)
plt.semilogx(np.linspace(0,fs/2-1,len(Rates_Median)),Rates,'g',linewidth=0.25)

plt.show()
```
